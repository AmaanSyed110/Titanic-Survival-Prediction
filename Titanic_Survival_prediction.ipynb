### **The aim of this project is to build a model that predict the survival rate of passengers in Titanic**

```python
#importing the required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
#importing the dataset and displaying the top 5 entries
df=pd.read_csv('/content/drive/MyDrive/Datasets/tested.csv')
df.head()
```

```python
#information about the datset
df.info()
```

```python
#statistical description of the dataset
df.describe()
```

```python
#dropping the 'Cabin' column
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
#Checking for null values
df.isna().sum()
```

```python
#Dropping the null values
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#Converting the "Sex" column to numeric
df['Sex'] = df['Sex'].apply(lambda x:1 if x=='male' else 0)
df.head()
```

```python
#Creating the feature and target variable
X=df.drop(['Survived','Name','Ticket','Embarked'],axis=1)
y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

## **Logistic Regression Model**

```python
model=LogisticRegression(max_iter=500)
model.fit(X_train,y_train)
```

```python
lr_pred=model.predict(X_test)
```

```python
lr_cr=classification_report(y_test,lr_pred)
print(lr_cr)
```

## **Random Forest Classifier Model**

```python
rfc_model=RandomForestClassifier()
```

```python
rfc_model.fit(X_train,y_train)
```

```python
rfc_model.score(X_test,y_test)
```

```python
pred=rfc_model.predict(X_test)
pred
```

```python
cr=classification_report(y_test,pred)
print(cr)
```

## **Decision Tree Classifier Model**

```python
dtc_model=DecisionTreeClassifier()
```

```python
dtc_model.fit(X_train,y_train)
```

```python
dtc_pred=dtc_model.predict(X_test)
```

```python
dtc_cr=classification_report(y_test,dtc_pred)
```

```python
print(dtc_cr)
```
